```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

path1 = 'images.jpg'

img = cv.imread(path1)

plt.imshow(img)
plt.show()
print(img)


# print(apply_kernel(img1, kernel_horizontal))
```

Blurring

```python
# Box Blurr
kernel_key = np.array([[0.1111111, 0.1111111, 0.1111111], 
                    [0.1111111, 0.1111111, 0.1111111], 
                    [0.1111111, 0.1111111, 0.1111111]])
kernel_key2 = np.array([[0.1111111, 0.1111111, 0.1111111, 0.1111111, 0.1111111 ], 
                    [0.1111111, 0.1111111, 0.1111111, 0.1111111, 0.1111111], 
                    [0.1111111, 0.1111111, 0.1111111, 0.1111111, 0.1111111],
                    [0.1111111, 0.1111111, 0.1111111, 0.1111111, 0.1111111], 
                    [0.1111111, 0.1111111, 0.1111111, 0.1111111, 0.1111111]])

kernel_key3 = np.array([[0.1111111, 0.1111111, 0.1111111, 0.1111111, 0.1111111 ], 
                    [0, 0.1111111, 0.1111111, 0.1111111, 0], 
                    [0, 0, 0.1111111, 0, 0],
                    [0, 0.1111111, 0.1111111, 0.1111111, 0], 
                    [0.1111111, 0.1111111, 0.1111111, 0.1111111, 0.1111111]])

Box_blur = cv.filter2D(src=img, ddepth=-1, kernel=kernel_key)
Box_blur2 = cv.filter2D(src=img, ddepth=-1, kernel=kernel_key2)
Box_blur3 = cv.filter2D(src=img, ddepth=-1, kernel=kernel_key3)

plt.imshow(img)
plt.show()
print(img)

plt.imshow(Box_blur)
plt.show()
print(Box_blur)

plt.imshow(Box_blur2)
plt.show()
print(Box_blur2)


plt.imshow(Box_blur3)
plt.show()
print(Box_blur3)

```

```python
# Gaussian Blurr
# kernel_key = np.array([[0.1111111, 0.1111111, 0.1111111], 
#                     [0.1111111, 0.1111111, 0.1111111], 
#                     [0.1111111, 0.1111111, 0.1111111]])


img4 = cv.imread('chestboard.jpg')

Gaussian_blur = cv.GaussianBlur(src=img4, ksize=(3,3), sigmaX=0, sigmaY=0)

Gaussian_blur2 = cv.GaussianBlur(src=img4, ksize=(15,15), sigmaX=0, sigmaY=0)

plt.imshow(img4)
plt.show()
print(img4)


plt.imshow(Gaussian_blur)
plt.show()
print(Gaussian_blur)


plt.imshow(Gaussian_blur2)
plt.show()
print(Gaussian_blur2)
```

```python
plt.imshow(Gaussian_blur2)
plt.show()

a,b,c = Gaussian_blur.shape

mask = np.zeros((a, b, 3), dtype=np.uint8)

mask = cv.rectangle(mask, (0,0), (a//2, b), (255,255,255), -1)
out = np.where(mask==(255, 255, 255), img, Gaussian_blur2)

plt.imshow(out)
plt.show()
```

Sharpening

```python
# Sharpening
kernel_key = np.array([[0, -1, 0], 
                    [-1, 5, -1], 
                    [0, -1, 0]])

kernel_key4 = np.array([[0, 0, -1, 0, 0], 
                    [0, -3, -1, -3, 0], 
                    [1, 3, 5, 3, 1],
                    [0, -3, -1, -3, 0],
                    [0, 0, -1, 0, 0]])


Sharpening = cv.filter2D(src=img, ddepth=-1, kernel=kernel_key)

Sharpening2 = cv.filter2D(src=img, ddepth=-1, kernel=kernel_key4)

plt.imshow(img)
plt.show()
print(img)


plt.imshow(Sharpening)
plt.show()
print(Sharpening)

plt.imshow(Sharpening)
plt.show()
print(Sharpening)
```

Emboss

```python
#Emboss

emboss_kernel = np.array([[-1, 0, 0], 
                    [0, 0, 0], 
                    [0, 0, 1]]) 

emboss_img = cv.filter2D(src=img, ddepth=-1, kernel=emboss_kernel)

plt.imshow(img)
plt.show()


plt.imshow(emboss_img)
plt.show()
```

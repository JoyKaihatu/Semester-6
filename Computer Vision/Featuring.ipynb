```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = 'Shape1.png'

img = cv2.imread(path)

plt.axis('off')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img, cmap='gray')


print(img.shape)
```

```python
imgflt = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
imgflt = np.float32(imgflt)

dst = cv2.cornerHarris(imgflt, blockSize=2, ksize=3,k=0.1)

```

```python
dst = cv2.dilate(dst, None) 
img[dst > 0.01 * dst.max()] = [0, 255, 0] 
```

```python
plt.imshow(img)
```

```python
num_corner = np.sum(dst > 0.01 * dst.max())

print(num_corner)
```
